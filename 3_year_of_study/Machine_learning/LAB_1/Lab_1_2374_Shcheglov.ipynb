```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('Netflix.csv')
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']

for column in columns:
    print(f"Обработка атрибута: {column}")
    
    mean_value = data[column].mean()
    std_value = data[column].std()
    print(f"Среднее значение для {column}: {mean_value}")
    print(f"Стандартное отклонение для {column}: {std_value}")
    
    plt.figure(figsize=(10, 6))
    sns.histplot(data[column], kde=True)
    plt.title(f'Гистограмма для {column}')
    plt.show()
    
    missing_values = data[column].isnull().sum()
    print(f"Пропущенные значения для {column}: {missing_values}")
    
    if missing_values > 0:
        print(f"Заполнение пропущенных значений в {column} средним")
        data[column].fillna(mean_value, inplace=True)
    
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = ((data[column] < (Q1 - 1.5 * IQR)) | (data[column] > (Q3 + 1.5 * IQR))).sum()
    print(f"Количество выбросов в {column}: {outliers}")
    
    print("\n" + "="*40 + "\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('Netflix.csv')
data = data.drop(columns=['Date'])  

correlation_matrix = data.corr()

print("Корреляционная матрица:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Корреляционная матрица между атрибутами')
plt.show()

sns.pairplot(data)
plt.show()

high_corr = correlation_matrix[(correlation_matrix > 0.8) | (correlation_matrix < -0.8)]
print("Высококоррелированные атрибуты:")
print(high_corr)
```

В данном шаге мы построили корреляционную матрицу между атрибутами и нашли высококоррелированные атрибуты нашего датасета

Немного справвочного материала: корреляция показывает степень линейной взаимосвязи между двумя переменными. Значения корреляции находятся в диапазоне от -1 до 1. Значения, близкие к 1, означают сильную положительную корреляцию, близкие к -1 — сильную отрицательную корреляцию, а значения, близкие к 0, означают отсутствие корреляции

Из последней матрицы видно, что все атрибуты, помимо Volume(Кол-во проданных акций за день), высококоррелированные. Это не удивительно ведь мы изучаем разные пиковые цены акций одной компании за один день, и каждый атрибут очень сильно будет зависеть друг от друга( цена закрытия попросто не может не зависеть от цены открытия ), другого и не могло быть. => все значения высококоррелированные, а значение Volume вообще не имеет корреляции.

Вывод: Мы выяснили, что многие параметры предоставленные в выбранном датасете высококоррелированны и нашли средние показатели по каждому атрибуту.
