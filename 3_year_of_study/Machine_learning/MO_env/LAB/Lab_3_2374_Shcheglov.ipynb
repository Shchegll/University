В программе вы присваеваем каждой сртоке новый показатель и находим количество ростов и падений.
Для сбалансированности нужно чтобы кол-во показателей одного вида было относительно схожим с кол-вом показателей другой формы, такое соотношение нужно чтобы были более точные результаты. Нам повезло, баланс классов примерно равный, с небольшим первесом в сторону падений цены за день

   Далее мы оцениваем нашу класификацию с помощью следующих алгоритмов:
   - kNN - алгоритм основанные на сходстве данных
   - Дерево решений - алгоритм основан на постройке дерева где на каждом узле выбирается атрибут для разбиения
   А для сравнения результатов воспользуемся метриками:
   - Accuracy - показывает долю правильно предсказанных полож объектов среди всех предсказанных
   - Presicion - показывает долю правильно предсказанных полож объектов среди всех реальных положительных объектов
   - Recall - улавливает все положительные случаи
   - F-measure - что-то среднее между Presicion и Recall, важно когда нужно сбалансировать оба показателя
 
 
 В нашем примере для kNN точность класификации составила 49% , но нельзя операться только на точность(Accuracy), ведь она может давать сбои, когда кластеры не несбалансированы( один класс повторяется многократно), для этого посмотрим на другие показатели. ( для пояснения в выводе программы support - это реальное кол-во таких случаев, macro avg - это среднее всех показателей, а weighted avg - взвешенное среднее, для большей точности. 
 Для дерева решений точность будет 73%, а остальные дополнерия я описал выше.

 Исходя из этих показателей, мы видим, что разница есть, что указывает на несбалансированность нашего примера, это может происходить из повторов данных.

 Также нам нужно построить ROC - кривую, что я продемонстрировал далее. Так более наглядно видно разницу
 Всё это говорит, что модель (дерево решений) лучше предсказыввает классы и имеет неплохие характеристики!
 

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score, roc_curve
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

data_cleaned = pd.read_csv('/Users/yarik_sglv/University/3_year_of_study/Machine_learning/MO_env/LAB/Updated_Netflix.csv')

data_cleaned['Price Direction'] = (data_cleaned['Close'] > data_cleaned['Open']).astype(int)
class_distribution = data_cleaned['Price Direction'].value_counts()

# Выбираем целевые и входные признаки
X = data_cleaned[['Open', 'High', 'Low', 'Volume', 'Adj Close']]
y = data_cleaned['Price Direction']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

print("Accuracy (kNN):", accuracy_score(y_test, y_pred_knn))
print("\nClassification Report (kNN):\n", classification_report(y_test, y_pred_knn))





tree = DecisionTreeClassifier(random_state=42)
tree.fit(X_train, y_train)

y_pred_tree = tree.predict(X_test)

print("\nAccuracy (Decision Tree):", accuracy_score(y_test, y_pred_tree))
print("\nClassification Report (Decision Tree):\n", classification_report(y_test, y_pred_tree))





y_pred_knn_proba = knn.predict_proba(X_test)[:, 1]
roc_auc_knn = roc_auc_score(y_test, y_pred_knn_proba)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_pred_knn_proba)

y_pred_tree_proba = tree.predict_proba(X_test)[:, 1]
roc_auc_tree = roc_auc_score(y_test, y_pred_tree_proba)
fpr_tree, tpr_tree, _ = roc_curve(y_test, y_pred_tree_proba)

plt.figure(figsize=(10, 6))
plt.plot(fpr_knn, tpr_knn, label=f'kNN (AUC = {roc_auc_knn:.2f})')
plt.plot(fpr_tree, tpr_tree, label=f'Decision Tree (AUC = {roc_auc_tree:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Диагональ для случайного угадывания
plt.title('ROC-кривая')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
